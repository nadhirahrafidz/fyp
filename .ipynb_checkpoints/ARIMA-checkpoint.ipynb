```python
# ALL 51 regions
symptoms = F[(F.date >= '2020-03-05') & (F.date <= '2020-12-31')].sort_values(by=['date', 'open_covid_region_code'], ascending=True)
symptoms_grouped = symptoms.groupby(['open_covid_region_code', 'date']).sum()


test_symptoms = F_2021[(F_2021.date >= '2021-01-01') & (F_2021.date <= '2021-02-05')].sort_values(by=['date', 'open_covid_region_code'], ascending=True)
grouped_test_symp = test_symptoms.groupby(['open_covid_region_code', 'date']).sum()
```

```python
def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    me = np.mean(forecast - actual)             # ME
    mae = np.mean(np.abs(forecast - actual))    # MAE
    mpe = np.mean((forecast - actual)/actual)   # MPE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    corr = np.corrcoef(forecast, actual)[0,1]   # corr
    mins = np.amin(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    maxs = np.amax(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    minmax = 1 - np.mean(mins/maxs)             # minmax
    return({'mape':mape, 'me':me, 'mae': mae, 
            'mpe': mpe, 'rmse':rmse,
            'corr':corr, 'minmax':minmax})

forecast_accuracy(pred, Y)
```
